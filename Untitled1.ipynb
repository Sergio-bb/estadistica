```python
# librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# carga la dataset
df = pd.read_csv('/content/content/train.csv')
```

1. Neighborhood \
  a. ¿Cuáles son los vecindarios más comunes en el conjunto de datos?\
      Como se puede observar en la grafica, el barrio mas comun es el NAmes

```python
df.head()

#Variable que almacena el conteo de la cantidad de registros por barrio
frecuencys = df['Neighborhood'].value_counts()
#vaariable que obtiene los dies barrios con mas cantidad de registros
top_10_neighborhood = frecuencys.head(10)




plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_neighborhood.values, y=top_10_neighborhood.index, palette="viridis")


plt.title('Top 10 Barrios Más Comunes', fontsize=15)
plt.xlabel('Número de Apariciones', fontsize=12)
plt.ylabel('Barrio', fontsize=12)

# Mostrar el gráfico
plt.show()

```

1. Neighborhood \
  b. ¿Existe una diferencia notable en el precio promedio de venta de casas entre los
diferentes vecindarios?\
      Como se puede observar en la grafica, La diferencia del precio promedio por vecindarios es notable.

```python
# Calcular el precio promedio por vecindario
precio_promedio = df.groupby('Neighborhood')['SalePrice'].mean().reset_index()

# Ordenar los vecindarios por el precio promedio
precio_promedio = precio_promedio.sort_values(by='SalePrice', ascending=False)

# Crear el gráfico de barras
plt.figure(figsize=(12, 8))
sns.barplot(x='SalePrice', y='Neighborhood', data=precio_promedio, palette='viridis')

# Agregar título y etiquetas
plt.title('Precio Promedio de Venta de Casas por Vecindario', fontsize=15)
plt.xlabel('Precio Promedio', fontsize=12)
plt.ylabel('Vecindario', fontsize=12)

# Mostrar el gráfico
plt.show()
```
